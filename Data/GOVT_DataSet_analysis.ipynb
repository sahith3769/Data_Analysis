# Dataset: NYC-2022-2023_0_0.csv


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Loading and Exploration
In this section, we load the dataset and perform an initial exploration to understand the structure and key statistics.


```python
df = pd.read_csv("NYC-2022-2023_0_0.csv", encoding="ISO-8859-1")
df
```

# National Youth Corps (NYC) Dataset Analysis (2016-19)

## Introduction
This notebook analyzes the National Youth Corps (NYC) dataset provided by the Ministry of Youth Affairs & Sports. 
The dataset covers:
- The number of applications received.
- The number of NYVs (National Youth Volunteers) selected.
- The category-wise selection of SC, ST, and OBC candidates (2016-17 to 2018-19).

The goal of this analysis is to understand selection trends, category representation, and overall recruitment patterns.


```python
df.columns = df.columns.str.strip()
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

# Display basic information

```python
display(df.head())
print(df.info())
print(df.describe())
```

# Check for missing values

```python
missing_values = df.isnull().sum()
print("Missing values per column:\n", missing_values)
```

```python
df.State.value_counts()
```

```python
df.iloc[:, 1:] = df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')
df.dropna(inplace=True)
```

```python
print("Numeric columns:",df.columns)
```

## Year-wise Trend of Applications and Selections
This section analyzes how the number of applications received and NYVs selected has changed over the years.
A comparison helps identify:
- Growth in applications.
- Variations in selection rates.


# Pairplot for numerical features

```python
sns.pairplot(df.iloc[:, 1:])
plt.show()
```

# Correlation heatmap

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.iloc[:, 1:].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

# Boxplot for gender-wise distribution

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df[['Male Member', 'Female Member']])
plt.title("Boxplot of Male and Female Members Across Districts")
plt.show()
```

# Violin plot for education levels

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df)
plt.title("Distribution of Education Levels Across Districts")
plt.show()
```

## Year-wise Trend of Applications and Selections
This section analyzes how the number of applications received and NYVs selected has changed over the years.
A comparison helps identify:
- Growth in applications.
- Variations in selection rates.


```python
plt.figure(figsize=(10, 5))
sns.barplot(x='State', y='Male Member', data=df, color='blue', label='Male')
sns.barplot(x='State', y='Female Member', data=df, color='pink', label='Female', bottom=df['Male Member'])
plt.xticks(rotation=90)
plt.ylabel("Number of Members")
plt.title("Gender Distribution Across Districts")
plt.legend()
plt.show()
```

# Pie chart for gender distribution

```python
plt.figure(figsize=(8, 8))
gender_counts = [df['Male Member'].sum(), df['Female Member'].sum()]
labels = ['Male Members', 'Female Members']
colors = ['green', 'yellow']
plt.pie(gender_counts, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
plt.title("Gender Distribution Pie Chart")
plt.show()
```

## Category-wise Selection Analysis
We analyze the number of selected NYVs based on social categories (SC, ST, and OBC) to understand diversity in recruitment.


```python
caste_cols = ['General', 'SC', 'ST', 'OBC']
df_caste = df.melt(id_vars=['State'], value_vars=caste_cols,var_name='Caste', value_name='Count')
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='Count', hue='Caste', data=df_caste)
plt.xticks(rotation=90)
plt.ylabel("Number of Members")
plt.title("Caste Category Distribution Across Districts")
plt.legend()
plt.show()
```

## Selection Ratio Analysis
Here, we calculate the percentage of applicants selected each year to determine the competitiveness of the selection process.


```python
edu_cols = ['10th', '12th', 'Graduate', 'Post Graduate']
df_edu = df.melt(id_vars=['State'], value_vars=edu_cols, 
                  var_name='Education Level', value_name='Count')
```

# Education level analysis

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='Count', hue='Education Level', data=df_edu)
plt.xticks(rotation=90)
plt.ylabel("Number of Members")
plt.title("Education Level Distribution Across States")
plt.legend(title="Education Level")
plt.show()
```

# Histogram of Male & Female Members

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Male Member'], bins=20, kde=True)
plt.title("Distribution of Male Members")
plt.xlabel("Male Members")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Female Member'], bins=20, kde=True)
plt.title("Distribution of Female Members")
plt.xlabel("Female Members")
plt.ylabel("Frequency")
plt.show()
```

## Year-on-Year Growth Analysis
This section compares year-over-year changes in applications received and NYVs selected to analyze recruitment trends and policy impacts.


```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=df[edu_cols])
plt.title("Distribution of Education Levels Across States")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Male Member'], shade=True, color='red', label='Male')
sns.kdeplot(df['Female Member'], shade=True, color='green', label='Female')
plt.title("Density Plot of Male and Female Members")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
state_counts = df['State'].value_counts()
sns.barplot(x=state_counts.values, y=state_counts.index, palette="Blues_r")
plt.xlabel("Count")
plt.ylabel("State")
plt.title("State-Wise Count Distribution")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxenplot(data=df[[ 'Male Member', 'Female Member']])
plt.title("Boxen Plot for Outlier Detection")
plt.show()
```

```python
df['Total Members'] = df['Male Member'] + df['Female Member']
top_districts = df.sort_values(by='Total Members', ascending=False).head(5)
bottom_districts = df.sort_values(by='Total Members').head(5)
print("Top 5 States by Total Members:\n", top_districts[['State', 'Total Members']])
print("Bottom 5 States by Total Members:\n", bottom_districts[['State', 'Total Members']])
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Total Members'], bins=20, kde=True)
plt.title("Distribution of Total Members")
plt.xlabel("Total Members")
plt.ylabel("Frequency")
plt.show()
```
