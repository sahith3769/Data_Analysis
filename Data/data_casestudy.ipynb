# Case Study 1 on matplotlib
# DataAnalysis on Class Marks
# Data set : class_marks.csv



```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("class_marks.csv")
df
```

```python
df1=df.sort_values("Total",ascending=True)
df1
```

# TOTAL values are sorted in ascending order

```python
df1=df1.fillna(0)
```

```python
df1
```

# fill 0 in all null values

```python

df1=df1.astype("int64")
```

```python
df1
```

# convert all values into int type

```python
df1['Q1']=df1["Q1aM4"]+df1["Q1bM6"]
df1['Q2']=df1['Q2aM6']+df1['Q2bM4']
df1['Q3']=df1['Q3aM5']+df1['Q3bM5']
df1['Q4']=df1['Q4aM3']+df1['Q4bM7']
df1['Q5']=df1['Q5M10']
df1['Q6']=df1['Q6aM4']+df1['Q6bM6']
df1.drop(columns=["Q1aM4","Q1bM6","Q2aM6","Q2bM4","Q3aM5","Q3bM5","Q4aM3","Q4bM7","Q5M10","Q6aM4","Q6bM6"],axis=1,inplace=True)
df1
```

# New Columns Q1,Q2,Q3,Q4,Q6 are created above to do analysis.

```python
a=df1.loc[(df1['Total'] >= 15) & (df1['Total'] <= 20)]
a=a.reset_index()
a
```

# Total marks 15-20 are filtered

```python
h1=a.groupby('Q1')['Total']
h1.hist(color = 'yellow',figsize=[8,8],grid=False,bins=6)
plt.title("Histogram of students who scored 15-20 Marks")
plt.xlabel("Total Marks obtained")
plt.ylabel("Marks in Q1")
plt.show()
```

# Most of the students scored 1 mark and maximum mark is 2, implying that all the students in this range secured very less marks in Q1

```python
h2=a.groupby('Q2')['Total']
h2.hist(color = 'orange',figsize=[8,8],grid=False,bins=6)
plt.title("Histogram of students who scored 15-20 Marks")
plt.xlabel("Total Marks obtained")
plt.ylabel("Marks in Q2")
plt.show()
```

# Very few students got high marks in Q2 and that's only 2, so overall performance in Q2 is not upto the mark

```python
h3=a.groupby('Q3')['Total']
h3.hist(color = 'green',figsize=[8,8],grid=False,bins=6)
plt.title("Histogram of students who scored 15-20 Marks")
plt.xlabel("Total Marks obtained")
plt.ylabel("Marks in Q3")
plt.show()
```

# Only few students got 3 marks whereas others scored below 2 marks, implying very less marks are secured in Q3 overall

```python
h4=a.groupby('Q4')['Total']
h4.hist(color = 'maroon',figsize=[8,8],grid=False,bins=7)
plt.title("Histogram of students who scored 15-20 Marks")
plt.xlabel("Total Marks obtained")
plt.ylabel("Marks in Q4")
plt.show()
```

# Many of these students who attempted Q4 got more marks and maximum marks are 4 others got less than 2 marks

```python
h5=a.groupby('Q5')['Total']
h5.hist(color = 'blue',figsize=[8,8],grid=False,bins=6)
plt.title("Histogram of students who scored 15-20 Marks")
plt.xlabel("Total Marks obtained")
plt.ylabel("Marks in Q5")
plt.show()
```
