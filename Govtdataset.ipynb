# Data Cleaning
Datasets often contain missing or incorrect values. Here, we replace missing values with zero to ensure smooth analysis.

````python
df.drop_duplicates(inplace=True)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns


df_20 = df.head(20)


numeric_cols = [
    'During the current year upto the end of month under review',
    'During the corresponding month of previous year',
    'During the previous month',
    'During the current month'
]
df_20[numeric_cols] = df_20[numeric_cols].apply(pd.to_numeric, errors='coerce')


plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")


plt.figure(figsize=(12, 6))
sns.barplot(x=df_20['During the current month'], y=df_20['MAJOR HEADS'], palette="viridis")
plt.xlabel("Total Crimes (Current Month)")
plt.ylabel("Major Crime Heads")
plt.title("Crime Distribution by Major Heads (Current Month)")
plt.show()



````

````python
plt.figure(figsize=(10, 5))
df_20[numeric_cols].sum().plot(marker='o', linestyle='-', color='b')
plt.xlabel("Time Periods")
plt.ylabel("Total Crimes")
plt.title("Crime Trends Over Different Time Periods")
plt.show()
````

````python
plt.figure(figsize=(8, 8))
df_20.groupby('MAJOR HEADS')['During the current month'].sum().plot(kind='pie', autopct='%1.1f%%', cmap='coolwarm')
plt.ylabel("")
plt.title("Crime Distribution by Major Heads (Current Month)")
plt.show()
````

````python
plt.figure(figsize=(8, 6))
sns.heatmap(df_20[numeric_cols].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Crime Data")
plt.show()
````

````python
# Fixing KDE Plot issue by ensuring numeric values only and dropping NaN
plt.figure(figsize=(10, 5))
for col in numeric_cols:
    sns.kdeplot(df_20[col].dropna().astype(float), fill=True, label=col)
plt.xlabel("Crime Count")
plt.ylabel("Density")
plt.title("Kernel Density Estimate (KDE) of Crime Data")
plt.legend()
plt.show()


top_10 = df_20.nlargest(10, 'During the current month')
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10['During the current month'], y=top_10['MAJOR HEADS'], palette="plasma")
plt.xlabel("Total Crimes (Current Month)")
plt.ylabel("Major Crime Heads")

plt.title("Top 10 Major Crime Heads (Current Month)")
plt.show()
````

````python
plt.figure(figsize=(12, 6))
sns.boxenplot(data=df_20[numeric_cols], palette="coolwarm")
plt.xlabel("Time Periods")
plt.ylabel("Crime Count")
plt.title("Crime Distribution Over Different Time Periods (Boxen Plot)")
plt.show()
````

````python
major_head = df_20.iloc[0]['MAJOR HEADS']
values = df_20.iloc[0][numeric_cols].values
categories = numeric_cols

angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
values = np.append(values, values[0])  # Repeat first value to close the circle
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, values, color='red', alpha=0.25)
ax.plot(angles, values, color='red', linewidth=2)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, fontsize=10)
plt.title(f"Crime Distribution for {major_head} (Radar Plot)")
plt.show()
````
